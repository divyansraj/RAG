```python
from docx import Document as docx
import os
```

```python
from langchain_community.document_loaders import Docx2txtLoader, UnstructuredWordDocumentLoader
```

```python
##Method 1:Using Docx2txtLoader
print("Method 1: Using Docx2txtLoader")

try:
    docx_loader =Docx2txtLoader("./data/docx/proposal.docx")
    docx_docs = docx_loader.load()
    print(f"Number of documents loaded: {len(docx_docs)}")
    print(f"Content of the first document:\n{docx_docs[0].page_content[:500]}...")  # Print the first 500 characters
    print(f"Metadata of the first document:\n{docx_docs[0].metadata}")
except Exception as e:
    print(f"Error loading document with Docx2txtLoader: {e}")
```

```python
##Method 2:Using UnstructuredWordDocumentLoader
print("\nMethod 2: Using UnstructuredWordDocumentLoader")
try:
    unstructured_loader = UnstructuredWordDocumentLoader("./data/docx/proposal.docx", mode="elements")
    unstructured_docs = unstructured_loader.load()
    print(f"Number of documents loaded: {len(unstructured_docs)}")
    for i,doc in enumerate(unstructured_docs[:3]):  # Print the first 3 documents
        print(f"\nElement {i+1}:")
        print(f"Type: {doc.metadata.get('category', 'unknown')}")
        print(f"Content:\n{doc.page_content[:500]}...")  # Print the first 500 characters
except Exception as e:
    print(f"Error loading document with UnstructuredWordDocumentLoader: {e}")
```
